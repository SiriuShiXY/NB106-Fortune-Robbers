# Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from random import sample
```

# 1. Tianchi Dataset

```python
dataset_1 = pd.read_excel("/Users/liyuanheng/Desktop/金融评论情绪分类/Data/1. Tianchi.xlsx")

comment_lst_1 = dataset_1["title"].tolist()
label_1 = dataset_1["emotion"].tolist()

dataset_1
```

```python
label_counts = pd.Series(label_1).value_counts()

plt.figure(figsize=(10, 6))
label_counts.plot(kind='bar')
plt.title('Distribution of Labels')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()

label_counts
```

```python
comment_lengths = [len(comment) for comment in comment_lst_1]

plt.figure(figsize=(10, 6))
plt.hist(comment_lengths, bins=50, color='skyblue', edgecolor='black')
plt.xlabel('Comment Length')
plt.ylabel('Frequency')
plt.title('Distribution of Comment Lengths')
plt.show()
```

# 2. CSDN中文股票评论文本数据集

```python
dataset_2 = pd.read_csv("/Users/liyuanheng/Desktop/金融评论情绪分类/Data/2. CSDN中文股票评论文本数据集.csv")

comment_lst_2 = dataset_2["sentence"].tolist()
label_2 = dataset_2["label"].tolist()

dataset_2
```
